##### **Graph 3**

```python
dfc= dfp.loc[dfp['GEO'].isin(['Vancouver International, British Columbia','Québec/Jean Lesage International, Quebec','Edmonton International, Alberta','Calgary International, Alberta','Halifax/Robert L Stanfield International, Nova Scotia','Winnipeg/James Armstrong Richardson International, Manitoba','Saskatoon/John G Diefenbaker International, Saskatchewan','Montréal/Pierre Elliott Trudeau International, Quebec','Toronto/Lester B Pearson International, Ontario','Ottawa/Macdonald-Cartier International, Ontario'])]
ax = sns.lmplot(data=dfc, x="REF_DATE",y="VALUE",hue="GEO")
ax.fig.set_size_inches(15,10)
ax.fig.suptitle("Yearly Passenger Count For Famous 10 Candian Cities for tourists")
plt.show()
```

##### **Graph 3 Analysis**

Based on the plot of the yearly passenger count for the famous cities over Canada, trendlines show that almost all the cities are steadily on a net incline, with Toronto airport having a much steeper slope compared to other airports, even though the pandemic caused the declining in 2020. We can then assume that before the pandemic, the general public is becoming more dependent on traveling through planes. However, the dataset may seem monotonous it is only provided with passenger count in a given year and its locations but not a survey based on true reflection on how travelers feel about the convenience of air travel. Over the decades, Canadian cities are gradually becoming hotspots for domestic and international travelers, traveling by plane will likely become more popular.
